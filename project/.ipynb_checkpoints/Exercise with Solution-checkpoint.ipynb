# Part 2: Linear Regression Analysis

## Choose one predictor variable and perform a simple linear regression analysis with median home value as the response variable.

*  For a simple linear regression, we typically start with the variable that has the strongest correlation.
* Let's perform a simple linear regression using rm as the predictor variable and medv as the response variable

```python
import statsmodels.api as sm
```

```python
# Define the predictor variable (RM) and the response variable (MEDV)
X = df['rm']  # Predictor
y = df['medv']  # Response
```

```python
# Add a constant to the predictor variable to include an intercept in the model
X_with_constant = sm.add_constant(X)

# Fit the simple linear regression model
simple_model = sm.OLS(y, X_with_constant).fit()

# Get the summary of the regression
simple_model_summary = simple_model.summary()
simple_model_summary
```

## Interpret the coefficients and calculate the coefficient of determination (R2).

* The regression equation is medv = 34.6706 + 9.1021  rm
* The R2 value is 0.484, indicating that approximately 48.4% of the variability in median home value can be explained by the number of rooms per dwelling.
* The p-value for rm is less than 0.001, indicating that the number of rooms is a statistically significant predictor of median home value.

```python
residuals = simple_model.resid

# Plotting the residuals to check for patterns
plt.figure(figsize=(12, 5))

# Residuals vs Fitted Values
plt.subplot(1, 2, 1)
sns.scatterplot(x=simple_model.fittedvalues, y=residuals)
plt.title('Residuals vs Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='red', linestyle='--')

# Q-Q plot for normality of residuals
plt.subplot(1, 2, 2)
sm.qqplot(residuals, line='s', ax=plt.gca())
plt.title('Q-Q Plot of Residuals')

plt.tight_layout()
plt.show()
```

## Conduct a multiple linear regression with a subset of variables based on your correlation analysis.

* Let's use both rm (average number of rooms per dwelling) and lstat (percent lower status of the population) for our multiple regression model since they had strong correlations with medv and represent different aspects of the housing data.

```python
# Create a pairplot for a subset of variables including 'medv', 'RM', and 'LSTAT'
sns.pairplot(df[['medv', 'rm', 'lstat']])
plt.suptitle('Pairplot of Median Home Values, Average Number of Rooms, and Percent Lower Status', 
             size=16, y=1.02)
plt.show()
```

```python
# Define the predictor variables and the response variable for multiple regression
X_multi = df[['rm', 'lstat']]  # Predictor variables
y_multi = df['medv']  # Response variable
```

```python
# Add a constant to the predictor variables to include an intercept in the model
X_multi_with_constant = sm.add_constant(X_multi)

# Fit the multiple linear regression model
multi_model = sm.OLS(y_multi, X_multi_with_constant).fit()

# Get the summary of the regression
multi_model_summary = multi_model.summary()
multi_model_summary
```

* The regression equation is medv=1.3583+5.0948rm0.6424lstat
* The R2 value is 0.639, meaning that approximately 63.9% of the variability in median home value can be explained by the number of rooms and the lower status population percentage.
* The p-values for both rm and lstat are less than 0.001, indicating that both predictors are statistically significant in explaining the median home value.

## Interpret the model coefficients and discuss any signs of multicollinearity.

* The condition number is a diagnostic metric used in regression analysis to assess the presence and severity of multicollinearity among the independent variables. Multicollinearity occurs when two or more independent variables in a regression model are highly linearly correlated. 
* Interpreting the Condition Number:
    - Low Condition Number: A low condition number (generally values less than 10 to 30) indicates that multicollinearity is not a concern, suggesting that the regression coefficients are estimated reliably.
    - High Condition Number: A high condition number (values significantly greater than 30, with higher thresholds like 100 or 1000 often cited) suggests that the matrix of predictors is close to singular, indicating a high level of multicollinearity. This means that small changes in the data or in the model specification could lead to large changes in the coefficient estimates, making them unstable and unreliable.
* The condition number is reported to be 202, which does not indicate severe multicollinearity; however, if this number were much larger, it could suggest potential problems.

## Check the normality of the residuals using a Q-Q plot.

## Assess homoscedasticity by plotting the residuals versus the fitted values.

```python
# Calculate the residuals for the multiple regression model
residuals_multi = multi_model.resid
```

```python
# Calculate the residuals for the multiple regression model
residuals_multi = multi_model.resid

# Plotting the residuals to check for patterns in the multiple regression model
plt.figure(figsize=(12, 5))

# Residuals vs Fitted Values
plt.subplot(1, 2, 1)
sns.scatterplot(x=multi_model.fittedvalues, y=residuals_multi)
plt.title('Residuals vs Fitted Values (Multiple Regression)')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='red', linestyle='--')

# Q-Q plot for normality of residuals
plt.subplot(1, 2, 2)
sm.qqplot(residuals_multi, line='s', ax=plt.gca())
plt.title('Q-Q Plot of Residuals (Multiple Regression)')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import shapiro
shapiro(residuals_multi)
```

## Discuss any patterns you find and what they might mean for the validity of your model.

* This plot shows a bit more of a random spread compared to the simple linear regression, suggesting an improvement in the homoscedasticity assumption, though there may still be some patterns to explore.
* The Q-Q plot for the multiple regression appears to show a better fit to the normal distribution than the simple regression, especially at the tails, but there is still some deviation (The normality assumption is still not met).

# Part 3: Interpetation of Results

## Calculate R2, MAE, and MSE for your models.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# For the simple linear regression model
simple_model_fitted_values = simple_model.fittedvalues
simple_model_r2 = simple_model.rsquared
simple_model_mae = mean_absolute_error(y, simple_model_fitted_values)
simple_model_mse = mean_squared_error(y, simple_model_fitted_values)
```

```python
# For the multiple linear regression model
multi_model_fitted_values = multi_model.fittedvalues
multi_model_r2 = multi_model.rsquared
multi_model_mae = mean_absolute_error(y_multi, multi_model_fitted_values)
multi_model_mse = mean_squared_error(y_multi, multi_model_fitted_values)
```

```python
# Create a DataFrame to display the results
model_performance_df = pd.DataFrame({
    'Model': ['Simple Regression', 'Multiple Regression'],
    'R^2': [simple_model_r2, multi_model_r2],
    'MAE': [simple_model_mae, multi_model_mae],
    'MSE': [simple_model_mse, multi_model_mse]
})

model_performance_df
```

## Create bar plots to compare the R2, MAE, and MSE across the simple and multiple regression models.

```python
# Set up the matplotlib figure
fig, axes = plt.subplots(1, 3, figsize=(20, 5))
fig.suptitle('Performance Metrics Comparison Between Simple and Multiple Regression Models')

# R^2 Comparison
sns.barplot(x='Model', y='R^2', data=model_performance_df, ax=axes[0])
axes[0].set_title('R^2 Comparison')
#axes[0].set_ylim(0, 1)  # Set the limit to show up to 100% for R^2

# MAE Comparison
sns.barplot(x='Model', y='MAE', data=model_performance_df, ax=axes[1])
axes[1].set_title('MAE Comparison')

# MSE Comparison
sns.barplot(x='Model', y='MSE', data=model_performance_df, ax=axes[2])
axes[2].set_title('MSE Comparison')

# Display the plots
plt.tight_layout()
plt.show()
```

* The R2 plot shows that the multiple regression model has a higher R2 value, indicating a better fit to the data.
* The MAE plot indicates that the multiple regression model has a lower mean absolute error, suggesting it has better average accuracy in predicting the median home values.
* The MSE plot also shows a lower value for the multiple regression model, which means it has a lower mean squared error and thus better predictive accuracy.
