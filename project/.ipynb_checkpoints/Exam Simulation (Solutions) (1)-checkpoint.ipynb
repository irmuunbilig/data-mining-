### Extract the Title from the Name column 
**Use the following mapping:**
* 'Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona' --> 'Noble'
* 'Mme','Ms' --> 'Mrs'
* 'Mlle' --> 'Miss'
* 'Mr'
* 'Master'

```python
def get_title(name):
    # Split the name by space to get individual words
    words = name.split()

    # Iterate through words to find a title
    for word in words:
        # Check if the word ends with a period and contains only alphabetical characters
        if word.endswith('.') and word[:-1].isalpha():
            # Return the word (title) without the period
            return word[:-1]

    # Return an empty string if no title is found
    return ""
```

```python
df_clean['Title'] = df['Name'].apply(get_title)
```

```python
df_clean['Title'].unique()
```

```python
def replace_titles(x):
    title = x['Title']
    if title in ['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona']:
        return 'Noble'
    elif title in ['Mme','Ms']:
        return 'Mrs'
    elif title in ['Mlle']:
        return 'Miss'
    else:
        return title
```

```python
df_clean['Title'] = df_clean.apply(replace_titles, axis=1)
```

```python
df_clean['Title'].unique()
```

```python
# Check the survival rate for the title cathegories obtained
df_grouped = df_clean.groupby(['Title', 'Sex']).agg({'Survived': ['sum', lambda x: x.count() - x.sum()]})
df_grouped.columns = ['Survived', 'Non-Survived']
df_grouped['Count'] = df_clean.groupby(['Title', 'Sex']).size()
# Calculate the survived/non-survived rate
df_grouped['Survival Rate'] = df_grouped['Survived'] / (df_grouped['Survived'] + df_grouped['Non-Survived'])
# Order the DataFrame by the survival rate in descending order
df_grouped = df_grouped.sort_values('Survival Rate', ascending=False)

print(df_grouped)
```

### Turning cabin number into Deck
* The letter refers to the deck (e.g., C85 -> C)
* The missing values can be imputed by creating a new categorical variable N

```python
import re 
pattern = r'([A-Za-z])' # It considers only the letter
df_clean['Deck'] = df_clean['Cabin'].fillna('N').apply(lambda x: re.search(pattern, str(x)).group(1) if x != 'N' else 'N')
df_clean.drop(columns = ['Cabin'], inplace = True)
```

```python
df_clean
```

### Map the 'Age' variable
* 013 -> kids 
* 1330 -> young
* 3045 -> adult
* 4560 -> old
* older than 60 years -> very old

```python
# Mapping Age
df_clean.loc[df_clean['Age'] <= 13, 'Age_group'] = 0 # kids
df_clean.loc[(df_clean['Age'] > 13) & (df_clean['Age'] <= 30), 'Age_group'] = 1 # young
df_clean.loc[(df_clean['Age'] > 30) & (df_clean['Age'] <= 45), 'Age_group'] = 2 # adult
df_clean.loc[(df_clean['Age'] > 45) & (df_clean['Age'] <= 60), 'Age_group'] = 3 # old
df_clean.loc[(df_clean['Age'] > 60) & (df_clean['Age'] <= 100), 'Age_group'] = 4 # very old
```

```python
# Calculate survival probability for each fare level
age_survival_grouped = df_clean.groupby('Age_group').agg({'Age': ['min', 'max'], 'Survived': ['mean', 'count']}).reset_index()
age_survival_grouped.columns = ['Age_group', 'Age_min' , 'Age_max', 'Survival_proba', 'Count']
age_survival_grouped
```

### Impute missing values using interpolation 
* Interpolation methods --> they estimate missing values based on the values of neighbouring data points.
* https://scikit-learn.org/stable/modules/generated/sklearn.impute.KNNImputer.html

```python
from sklearn.impute import KNNImputer

# Initialize the KNNImputer
knn_imputer = KNNImputer(n_neighbors=5)

# Perform imputation using KNN
df_clean[['Age', 'Age_group']] = knn_imputer.fit_transform(df_clean[['Age', 'Age_group']])
```

```python
df_clean['Age_group'] = df_clean['Age_group'].astype(int)
```

## 7) Data Splitting

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
X = df_clean.drop('Survived',axis=1)
y = df_clean['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## 8) Model Development

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Define the Random Forest classifier
rf_classifier = RandomForestClassifier()
# Define the hyperparameter grid for tuning
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
# Get the best model
best_rf_model = grid_search.best_estimator_
```

```python
# Evaluate the model on the test set
y_pred = best_rf_model.predict(X_test)
```

## 9) Model Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Evaluation
conf_matrix = confusion_matrix(y_test, y_pred)

print(classification_report(y_test, y_pred))

plt.figure(figsize=(4,2))
sns.heatmap(pd.DataFrame(conf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.tight_layout()
plt.title('Confusion matrix', y=1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```
