# Model Development in Supervised Learning

# 1) Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# 2) Loading the Dataset

```python
file_path = 'breastcancer_df.csv'  
df = pd.read_csv(file_path)
```

# 3) Data Exploration

* Data Type

# 4)  Preprocessing the Data
* Train test split
* Standardization

```python
X = df.drop('diagnosis', axis=1) 
y = df['diagnosis']
```

```python
# Data Standardization
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# 5) Model Training and Evaluation
* Logistic Regression
* SVM
* Decision Tree
* Random Forest

```python
models = {
    'Logistic Regression': LogisticRegression(),
    'SVM': SVC(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier()
}
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {accuracy * 100:.2f}%")
    print(f"Classification Report for {name}:\n", classification_report(y_test, y_pred))
    print('\n')
```

# 6) Hyperparameter Tuning 

```python
param_grid = {'C': [0.1, 1, 10], 'gamma': [1, 0.1, 0.01], 'kernel': ['rbf', 'poly', 'sigmoid']}
grid_search = GridSearchCV(SVC(), param_grid, refit=True, verbose=2, cv=5)
grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_
```

```python
print("Best parameters for SVM:", best_params)
```

```python
best_model = grid_search.best_estimator_
predictions =  best_model.predict(X_test)
```

```python
conf_matrix = confusion_matrix(y_test, predictions)

plt.figure(figsize=(4,2))
sns.heatmap(pd.DataFrame(conf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.tight_layout()
plt.title('Confusion matrix', y=1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

```python
print(classification_report(y_test, predictions))
```

# 7) Kmeans 

### Use the Kmeans algorithm to create k cluster. Use the Silhouette method to try different k (2-10) and find out which one is the best

```python
from sklearn.cluster import KMeans                
from sklearn import metrics  

k_to_test = range(2,11,1) 
silhouette_scores = []

for k in k_to_test:
    model_kmeans_k = KMeans(n_clusters = k, n_init=10, random_state=0)
    model_kmeans_k.fit(X_scaled)
    labels_k = model_kmeans_k.labels_
    score_k = metrics.silhouette_score(X_scaled, labels_k)
    silhouette_scores.append(score_k)
    print("k = %d\tSI: %5.4f" % (k, score_k))
```

```python
plt.figure(figsize = (16,5))
plt.plot(range(2,11,1),silhouette_scores)
plt.show()
```

**2 provide the best score**

### Apply the Kmeans algorithm using the optimal k value. Then, extract the labels and centroids for each cluster.

```python
kmeans_model = KMeans(n_clusters = 2,random_state=0,n_init=10)
kmeans_model.fit(X_scaled) 
```

```python
labels = kmeans_model.predict(X_scaled)
```

```python
centroids = scaler.inverse_transform(kmeans_model.cluster_centers_)
```

```python
print('Centroids 1 cluster:',centroids[0])
print('\n Centroids 2 cluster:',centroids[1])
```

```python
values = {'M':1,'B':0}
y = y.replace(values)
```

```python
accuracy = accuracy_score(y, subset['Cluster'])
```
