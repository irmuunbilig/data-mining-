## 6) Outliers 

### Detect outliers in the "Age" column using both box plots and the IQR method

```python
plt.figure(figsize=(10,8))

df.boxplot(column='Age',grid=False,vert=False)
plt.show()
```

```python
df['Age'].hist()
plt.show()
```

```python
def detect_outliers(df,feature_name):
    
    Q1 = df[feature_name].quantile(0.25)
    Q3 = df[feature_name].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lwr_bound = Q1 -1.5 * IQR
    upp_bound = Q3 +1.5 * IQR 
    
    ls = df.index[np.logical_or(df[feature_name]<lwr_bound, 
                                df[feature_name]>upp_bound)] 
    return ls
```

```python
out = detect_outliers(cleaned_df,'Age')
```

```python
out.shape
```

```python
print("Outliers:")
print(out)
```

### Decide on an appropriate treatment strategy for outliers

```python
cleaned_df = cleaned_df.replace(cleaned_df['Age'].iloc[out].values,cleaned_df['Age'].median())
```
