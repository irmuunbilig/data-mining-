# Model Development

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df['Outcome'].value_counts()
```

```python
df.describe()
```

```python
# Addressing Zero values
from sklearn.impute import SimpleImputer

# Replacing 0 values with NaN in columns where 0 is not realistic
columns_with_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[columns_with_zeros] = df[columns_with_zeros].replace(0, np.nan)

# Imputing the missing values using median
imputer = SimpleImputer(strategy='median')
df[columns_with_zeros] = imputer.fit_transform(df[columns_with_zeros])

# Display the first few rows after imputation
df.describe()
```

```python
df.head()
```

```python
sns.set_theme(style='ticks')
sns.pairplot(df,diag_kind='hist',hue='Outcome',palette='muted')
plt.show()
```

# Data Splitting
* **Purpose**: To split the dataset into training and testing sets.
* **Method**: Use train_test_split from sklearn.model_selection.
* **Typical Split**: 70-80% for training, 20-30% for testing.

```python
from sklearn.preprocessing import StandardScaler

# Splitting the data into training and testing sets (70-30 split)
X = df.drop('Outcome', axis=1)
y = df['Outcome']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Model Training
* **Model Initialization**: Creating an instance of the model.
* **Model Fitting**: Training the model on the training data.
* **Models**:
    - Logistic Regression (LR): A linear model for classification.
    - Support Vector Machine (SVM): Effective in high-dimensional spaces.
    - Decision Tree (DT): A tree-like model of decisions.
    - K-Nearest Neighbors (KNN): Based on the closest training examples in the feature space.
    - Naive Bayes (NB): Based on applying Bayes theorem with strong independence assumptions.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

def print_model_evaluation(model, X_train, y_train, X_test, y_test, model_name):
    # Train and predict
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Metrics
    accuracy = accuracy_score(y_test, y_pred) * 100
    precision = precision_score(y_test, y_pred) * 100
    recall = recall_score(y_test, y_pred) * 100
    f1 = f1_score(y_test, y_pred) * 100

    # Classification report and confusion matrix
    class_report = classification_report(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    
    %matplotlib inline
    # Print results
    print(f"Model: {model_name}")
    print(f"Accuracy: {accuracy:.2f}%")
    print(f"Precision: {precision:.2f}%")
    print(f"Recall: {recall:.2f}%")
    print(f"F1 Score: {f1:.2f}%")
    print("\nClassification Report:")
    print(class_report)
    plt.figure(figsize=(4,2))
    sns.heatmap(pd.DataFrame(conf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
    plt.tight_layout()
    plt.title('Confusion matrix', y=1)
    plt.ylabel('Actual label')
    plt.xlabel('Predicted label')
    plt.show()
```

## Logistic Regression

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=0)

# Confirming the size of the splits
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
# Logistic Regression
lr_model = LogisticRegression(max_iter=1000)
print_model_evaluation(lr_model, X_train, y_train, X_test, y_test,'Logistic Regression')
```

## Support Vector Machine

```python
from sklearn.svm import SVC

# Support Vector Machine
svm_model = SVC()
print_model_evaluation(svm_model, X_train, y_train, X_test, y_test,'SVM')
```

## kNN

```python
from sklearn.neighbors import KNeighborsClassifier
# Elbow method in order to choose the best k
err_rate=[]
range_k = range(1,15,1)
for i in range_k:
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    err_rate.append(np.mean(pred_i != y_test))
plt.figure(figsize=(14,8))
plt.plot(range_k,err_rate,color='g',linestyle='dotted',marker='o',markerfacecolor='red',markersize='12')
plt.xlabel('k')
plt.ylabel('Error')
plt.show()
```

```python
# K-Nearest Neighbors
knn_model = KNeighborsClassifier(n_neighbors=10)
print_model_evaluation(knn_model, X_train, y_train, X_test, y_test,'knn')
```

## Decision Tree

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Decision Tree
dt_model = DecisionTreeClassifier()
print_model_evaluation(dt_model, X_train, y_train, X_test, y_test,'DT')
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

# Naive Bayes
nb_model = GaussianNB()
print_model_evaluation(nb_model, X_train, y_train, X_test, y_test,'Naive Bayes')
```

# K-Fold Cross Validation 

```python
from sklearn.model_selection import cross_val_score

# Function to perform cross-validation and return average accuracy
def perform_cross_validation(model, X, y, cv=5):
    cv_scores = cross_val_score(model, X, y, cv=cv)
    return cv_scores.mean()
```

```python
# Cross-validation for each model
lr_cv_accuracy = perform_cross_validation(lr_model, X_scaled, y)
svm_cv_accuracy = perform_cross_validation(svm_model, X_scaled, y)
dt_cv_accuracy = perform_cross_validation(dt_model, X, y)
knn_cv_accuracy = perform_cross_validation(knn_model, X_scaled, y)
nb_cv_accuracy = perform_cross_validation(nb_model, X, y)
```

```python
cv_results = {
    "Logistic Regression": lr_cv_accuracy,
    "Support Vector Machine": svm_cv_accuracy,
    "Decision Tree": dt_cv_accuracy,
    "K-Nearest Neighbors": knn_cv_accuracy,
    "Naive Bayes": nb_cv_accuracy
}

# Convert to percentage format and print
formatted_results = {model: f"{accuracy * 100:.2f}%" for model, accuracy in cv_results.items()}
formatted_results
```

# Hyperparameter Tuning
1) Train-Test Split: First, split your dataset into a training set and a test set. The test set will be used for the final evaluation of your model.
2) Train-Validation Split: Next, split the training set further into a smaller training set and a validation set. The validation set is used for model tuning and hyperparameter optimization.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
param_grid = {'C': [0.1,1, 10, 100], 'kernel': ['rbf'], 'gamma': [1,0.1,0.01,0.001,0.0001]} 
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
# Creating the SVM model
svm_model = SVC()
```

```python
# GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(svm_model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
# Best parameters
best_parameters = grid_search.best_params_
print("Best Parameters: ", best_parameters)
```

```python
# Training the model with the best parameters
best_model = grid_search.best_estimator_
```

```python
# Predictions
predictions = best_model.predict(X_test)

# Evaluation
conf_matrix = confusion_matrix(y_test, predictions)

print(classification_report(y_test, predictions))

plt.figure(figsize=(4,2))
sns.heatmap(pd.DataFrame(conf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.tight_layout()
plt.title('Confusion matrix', y=1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

# Ensemble Methods

## Random Forest
* A random forest is a meta estimator that fits a number of decision tree classifiers on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting.

https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Define the Random Forest classifier
rf_classifier = RandomForestClassifier()
```

```python
# Define the hyperparameter grid for tuning
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Create the GridSearchCV object with k-fold cross-validation 
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
# Fit the model to the training data
grid_search.fit(X_train, y_train)
```

```python
# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
# Get the best model
best_rf_model = grid_search.best_estimator_
```

```python
# Evaluate the model on the test set
y_pred = best_rf_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy)
```

```python
# Evaluation
conf_matrix = confusion_matrix(y_test, y_pred)

print(classification_report(y_test, y_pred))

plt.figure(figsize=(4,2))
sns.heatmap(pd.DataFrame(conf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.tight_layout()
plt.title('Confusion matrix', y=1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```
