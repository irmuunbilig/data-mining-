# Data Exploration and PCA


* The **Tips** dataset is a dataframe which represents some tipping data where one waiter recorded information about each tip he received over a period of a few months working in one restaurant.
* Information about the variables: the tip in dollars, the bill in dollars, the sex of the bill payer, whether there were smokers in the party, the day of the week, the time of day and the size of the party.

### Spearman Correlation Coefficient
* It is useful when we work with categorical variables because it works with the rank of the variables instead of their values.

```python
rho,p = stats.spearmanr(tips['sex'], tips['tip'])
print(rho)
```

```python
s = 'R= {}%'.format(round(rho*100,2))
# ScatterPlot with categorical variables
sns.set_theme(style="whitegrid", palette="muted")

plt.figure(figsize=(6,8))
# Draw a categorical scatterplot to show each observation
sns.swarmplot(data=tips, x="tip", y="sex",hue='time')

plt.text(8, 0, s, fontsize=12)

plt.show()
```

* Swarmplot is similar to the scatterplot, jet the locations of points are adjusted automatically to avoid overlap even if the jitter value is not applied. https://seaborn.pydata.org/examples/scatterplot_categorical.html

## 3) Data Preprocessing 
* PCA requires numerical data, so we need to handle the categorical variables (sex, smoker, day, and time) appropriately. 
* After the encoding, we'll standardize the data since PCA is affected by the scale of the data.

```python
# Checking for missing values
missing_values = tips.isnull().sum()
print(missing_values)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
categorical_columns = ['sex', 'smoker', 'day', 'time']
# Label encoding for categorical variables
label_encoder = LabelEncoder()
labeled_data = tips[categorical_columns].apply(label_encoder.fit_transform)
```

```python
labeled_df = pd.DataFrame(labeled_data, columns=categorical_columns)
labeled_df.index = tips.index
labeled_df = pd.concat([tips.drop(categorical_columns, axis=1), labeled_df], axis=1)
```

```python
labeled_df
```

```python
# Standardizing the data
scaler = StandardScaler()
scaled_labeled_data = scaler.fit_transform(labeled_df)
```

## 4) PCA
*  PCA is a technique used to emphasize variation and bring out strong patterns in a dataset. 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
components = pca.fit_transform(scaled_labeled_data)

# Explained Variance Ratio Plot
explained_variance = pca.explained_variance_ratio_
plt.figure(figsize=(8, 4))
plt.plot(range(1, 8), explained_variance)
plt.xticks(range(1, 8))
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio by Principal Component')
plt.show()
```

* The explained variance ratio tells you how much information (variance) can be attributed to each of the principal components.

```python
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}
labels
```

**Observations:**
Variance Explained:
* The PCA resulted in 7 principal components (PCs) due to the reduced number of columns from Label Encoding.
* The variance explained by each PC is as follows (approximate percentages):
   - PC1: 32.7%
   - PC2: 24.5%
   - PC3: 14.7%
   - ... and so on.

```python
# Plotting the first two principal components with Label Encoding
plt.figure(figsize=(10, 8))
plt.scatter(components[:, 0], components[:, 1])
plt.title('PCA with Label Encoding - First two principal components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

* The scatter plot visualizes the dataset using the first two principal components obtained from PCA, where categorical variables were processed using Label Encoding.
* Each point represents an observation in the dataset, now transformed into the two-dimensional space defined by the first two principal components.
* The x-axis is the first principal component (PC1), and the y-axis is the second principal component (PC2).
* This visualization is a useful way to explore and understand the dataset's structure in a lower-dimensional space. It can reveal patterns, clusters, or outliers that might not be apparent in the original high-dimensional space.

## PCA for dimensionality reduction
* IRIS dataset, which contains measurements of iris flowers (sepal length, sepal width, petal length, petal width) and the species of each iris (the target variable).

```python
from sklearn.datasets import load_iris

# Loading the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Displaying the first few rows of the dataframe
iris_df.head()
```

```python
# Excluding the target variable for PCA
X = iris_df.drop('species', axis=1)
```

```python
# Standardizing the feature data
X_scaled = scaler.fit_transform(X)
```

```python
# Applying PCA
pca_iris = PCA()
pca_iris_features = pca_iris.fit_transform(X_scaled)
```

```python
# Variance explained by each principal component
explained_variance_iris = pca_iris.explained_variance_ratio_

# Plotting the first two principal components
plt.figure(figsize=(10, 8))
scatter = plt.scatter(pca_iris_features[:, 0], pca_iris_features[:, 1], c=iris_df['species'], cmap='viridis')
plt.title('PCA of Iris Dataset - First two principal components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(*scatter.legend_elements(), title="Species")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(range(1, 5), explained_variance_iris)
plt.xticks(range(1,5))
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio by Principal Component')
plt.show()
```

**Observations:**
* The first two principal components account for over 95% of the total variance.
* This scatterplot demonstrates how PCA aids in distinguishing between different species. It appears that the first two principal components do a good job of separating the different iris species, especially between species 0 and the others.
