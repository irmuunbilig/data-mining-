### 5) Outliers

The main methods to detect outliers:
    
    1) Boxplot
    2) Inter Quantile Range (IQR) = Q3 - Q1
    


```python
# data import
iris = pd.read_csv('iris_with_missing.csv')
```

**1) Boxplot**

```python
plt.figure(figsize=(10, 8))
iris.boxplot()
plt.title('Box Plots for Iris Features')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

The 'sepal width' variable contains some outliers

```python
plt.figure(figsize=(8,4))
plt.xticks(size = 20)
plt.yticks(size = 15)

iris.boxplot(column = 'sepal width (cm)',grid=False,vert=False)
plt.show()
```

**2) IQR**
* The IQR is a measure of statistical dispersion and is calculated as the difference between the third quartile (Q3) and the first quartile (Q1). Mathematically, IQR = Q3 - Q1.
* To identify potential outliers, the lower bound and upper bound of the IQR range are defined. These bounds are calculated as follows:
   * Lower Bound = Q1 - 1.5 * IQR
   * Upper Bound = Q3 + 1.5 * IQR
* Data points that are less than the lower bound or greater than the upper bound are flagged as potential outliers.



```python
def detect_outliers(df,feature_name):
    
    Q1 = df[feature_name].quantile(0.25)
    Q3 = df[feature_name].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lwr_bound = Q1 -1.5 * IQR
    upp_bound = Q3 +1.5 * IQR 
    
    ls = df.index[np.logical_or(df[feature_name]<lwr_bound, 
                                df[feature_name]>upp_bound)] 
    return ls #return the indexes
```

```python
outliers = detect_outliers(iris, 'sepal width (cm)')

print("Outliers:")
print(outliers)
```

**3) Z-score**
* The Z-score, also known as the standard score or standardization, is a statistical measure that measures how many standard deviations an individual data point is away from the mean.
* Z = (X - μ) / σ
* Data points with Z-scores significantly greater or smaller than a certain threshold (e.g., 2 or 3) are considered potential outliers.

```python
from scipy.stats import zscore
```

```python
iris.isnull().sum()
```

```python
z_scores = np.abs(zscore(iris['sepal width (cm)'].dropna())) # The score is influenced by missing values
```

```python
# Set a threshold for outlier detection
threshold = 2.5  # You can adjust this threshold as needed
```

```python
# Find the indices of the outliers
outlier_indices = np.where(z_scores > threshold)
```

```python
print("Indices of the outliers:")
print(outlier_indices)
```

No outliers were detected with the z-score method

### 7) Feature Scaling
Datasets usually have features spanning different magnitudes, range and units. This can be a problem because some machine learning algorithms are highly sensitive to these features. To solve this problem, feature scaling can be implemented:
* Normalization 
* Standardization

```python
df3 = pd.DataFrame([
                   [180000, 110, 18.9, 1400], 
                   [360000, 905, 23.4, 1800], 
                   [230000, 230, 14.0, 1300], 
                   [60000, 450, 13.5, 1500]], 
    
                   columns=['Col A', 'Col B',
                            'Col C', 'Col D'])
```

```python
df3.boxplot(return_type = "axes")
plt.grid(False)
plt.show()
```

**1) Normalization**

```python
(df3-df3.min())/(df3.max()-df3.min())
```

```python
#Sklearn 
from sklearn.preprocessing import MinMaxScaler
```

```python
# Scaler definition
scaler = MinMaxScaler(feature_range = (0,1))
# Application of the scaler to our dataframe
df3_norm = scaler.fit_transform(df3) # array

# Convert it to df
name=df3.columns
df3_norm = pd.DataFrame(df3_norm,columns=name)
df3_norm
```

```python
df3_norm.boxplot(grid=False)
plt.show()
```

**2) Standardization**

X'=(X-µ)/σ

```python
(df3-df3.mean())/df3.std()
```

```python
#Sklearn
from sklearn.preprocessing import StandardScaler
```
