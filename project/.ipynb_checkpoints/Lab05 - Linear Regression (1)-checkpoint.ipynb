# Assumptions Check
* Linearity -> The relationship between the independent variables and the dependent variable should be linear.
* Independence -> Observations should be independent of each other. 
* Homoscedasticity -> The residuals (differences between observed and predicted values) should have constant variance. 
* Normality -> The residuals should be normally distributed. 
* No multicollinearity -> The independent variables should not be too highly correlated with each other.

```python
import statsmodels.api as sm

# Define the predictor variables and the response variable
X = data[['TV', 'PopDoc', 'GDP']]  # Predictor variables
y = data['LifeEx']  # Response variable

# Add a constant to the predictor variable set to represent the intercept
X_with_constant = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X_with_constant).fit() #Ordinary Least Squares

# Get the summary of the regression
model_summary = model.summary()
model_summary
```

### Linearity and Homoscedasticity

```python
# Calculate residuals
residuals = model.resid

# Calculate predicted values
model_fitted_y = model.fittedvalues
```

#### Residuals vs. Predicted Values:
*  To check for homoscedasticity and non-linearity.
* It provides a general overview of the model's fit across all predicted values.

```python
# Plotting the residuals
plt.figure(figsize=(10,6))

# Residuals vs Fitted
sns.residplot(x=model_fitted_y, y=y, lowess=True, 
              scatter_kws={'alpha': 0.5}, 
              line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})
plt.title('Residuals vs Fitted')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')

# Annotations for homoscedasticity
abs_resid = abs(residuals).sort_values(ascending=False)
for i in abs_resid.index[:3]:
    plt.annotate(i, xy=(model_fitted_y[i], residuals[i]))

plt.show()
```

#### Residuals vs. Individual Predictor Variables:
*  To check the assumption of linearity between each predictor and the response variable.
* By plotting the residuals against each predictor variable separately, you can identify issues that might not be apparent from the combined residual plot. For example, one predictor may have a non-linear relationship with the outcome that isn't evident in the overall residuals vs. fitted values plot.

```python
# Plotting the residuals to check for patterns
plt.figure(figsize=(18, 5))

# Residuals vs. Number of TVs
plt.subplot(1, 3, 1)
sns.scatterplot(x=data['TV'], y=residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuals vs. Number of TVs')

# Residuals vs. Population per Doctor
plt.subplot(1, 3, 2)
sns.scatterplot(x=data['PopDoc'], y=residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuals vs. Population per Doctor')

# Residuals vs. GDP
plt.subplot(1, 3, 3)
sns.scatterplot(x=data['GDP'], y=residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residuals vs. GDP')

plt.tight_layout()
plt.show()
```

**Ideal Case**
* Linearity: The residuals should be randomly dispersed around the horizontal line (y=0).
* Homoscedasticity: The spread of the residuals should be roughly constant across all levels of the independent variables. If the residuals fan out or form a funnel shape as the value of the predictor increases, it suggests that the variance of the residuals is not constant (heteroscedasticity).

**Observations:**
* From the plots, it seems there may be some non-linear patterns, especially with TV and GDP, suggesting that a simple linear model may not be the best fit for these data.
* The residuals against GDP seem to fan out, indicating potential heteroscedasticity.

### Normality

```python
from scipy.stats import shapiro
import scipy.stats as stats

# Perform the Shapiro-Wilk test on the model residuals
shapiro(residuals)
```

```python
# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# Histogram of the residuals
sns.histplot(residuals, kde=True, ax=ax[0])
ax[0].set_title('Histogram of Residuals')
ax[0].set_xlabel('Residuals')
ax[0].set_ylabel('Frequency')

# Q-Q plot of the residuals
stats.probplot(residuals, dist="norm", plot=ax[1])
ax[1].set_title('Q-Q Plot of Residuals')

plt.tight_layout()
plt.show()
```

**Observations:**
* The histogram and the shapiro test suggest that the residuals are normally distributed.
* The points fall along the 45-degree reference line quite well, which suggests that the residuals are normally distributed. There are some deviations at the ends, but these are not too severe.

### Independence
* The Durbin-Watson (DW) statistic is used to test for the presence of autocorrelation at lag 1 in the residuals from a linear regression analysis. Autocorrelation refers to the correlation of a time series with its own past and future values. 
* Range of Values: The DW statistic ranges from 0 to 4.
    - A value of 2 indicates no autocorrelation.
    - Values approaching 0 indicate positive autocorrelation.
    - Values approaching 4 indicate negative autocorrelation.

* In terms of the **Durbin-Watson statistic** from our model summary (1.995), it is very close to 2, which suggests there is no autocorrelation in the residuals, thus supporting the independence assumption.

## Multiple Regression Model
We will compare the results of the model with the untransformed variables extracted from the dataset against the models with the following transformations:
1) Log-transformed X and untransformed Y
2) Untransformed X and log-transformed Y
3) Log-transformed X and Y

```python
# Define the function to create models with different transformations
def fit_transformed_models(data, log_X=False, log_Y=False):
    # Select the predictors and response based on the transformation flags
    X = data[['TV', 'PopDoc', 'GDP']]
    y = data['LifeEx']
    
    if log_X:
        X = np.log(X)  # Log-transform X (add small value to avoid log(0))
    if log_Y:
        y = np.log(y)  # Log-transform Y (add small value to avoid log(0))
    
    # Add a constant to the predictor variable set to represent the intercept
    X_with_constant = sm.add_constant(X)
    
    # Fit the linear regression model
    model = sm.OLS(y, X_with_constant).fit()
    
    return model
```

#### 1) Model with Log-transformed X and untransformed Y

```python
model_log_X = fit_transformed_models(data, log_X=True, log_Y=False)
```

```python
model_log_X.summary()
```

#### 2) Model with Untransformed X and Log-transformed Y

```python
model_log_Y = fit_transformed_models(data, log_X=False, log_Y=True)
```

```python
model_log_Y.summary()
```

#### 3) Model with Log-transformed X and Y

```python
model_log_XY = fit_transformed_models(data, log_X=True, log_Y=True)
```

```python
model_log_XY.summary()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
# Define a function to calculate MAE and MSE
def calculate_errors(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    r2 = r2_score(y_true,y_pred)
    return mae, mse,r2
```

```python
# Calculate errors for the original model
original_mae, original_mse, original_r2 = calculate_errors(y, model.fittedvalues)

# Calculate errors for model with log-transformed X
model_log_X_mae, model_log_X_mse,model_log_X_r2 = calculate_errors(y, model_log_X.fittedvalues)

# Calculate errors for model with log-transformed Y
y_log = np.log(data['LifeEx'] + 1e-6)  # Log-transformed Y
model_log_Y_mae, model_log_Y_mse,model_log_Y_r2 = calculate_errors(y_log, model_log_Y.fittedvalues)

# Calculate errors for model with log-transformed X and Y
model_log_XY_mae, model_log_XY_mse,model_log_XY_r2 = calculate_errors(y_log, model_log_XY.fittedvalues)
```

```python
# Create a DataFrame to summarize the MAE and MSE for all models
error_summary = pd.DataFrame({
    'Model': ['Original', 'Log X', 'Log Y', 'Log X and Y'],
    'r2 score': [original_r2,model_log_X_r2,model_log_Y_r2,model_log_XY_r2],
    'MAE': [original_mae, model_log_X_mae, model_log_Y_mae, model_log_XY_mae],
    'MSE': [original_mse, model_log_X_mse, model_log_Y_mse, model_log_XY_mse]
})

error_summary
```

**Observations:**
* The model with log-transformed X (Log X) has the highest R2 value and the lowest MAE, indicating that it provides the best fit and the most accurate predictions among the four models.
* The model with log-transformed Y (Log Y) shows the least performance in terms of R2 and has the highest MAE, indicating that transforming only the dependent variable did not improve the model.
* The model with both X and Y log-transformed (Log X & Y) shows an improvement over the original model but not as much as the Log X model.

##Residuals vs Fitted

```python
# Plotting the residuals of the model with log-transformed X

log_X_residuals = model_log_X.resid

# Calculate fitted values for the Log X model
log_X_fitted = model_log_X.fittedvalues

# Residuals vs. Fitted Plot
plt.figure(figsize=(10, 6))
sns.residplot(x=log_X_fitted, y=y, lowess=True, 
              scatter_kws={'alpha': 0.5}, 
              line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})
plt.title('Residuals vs Fitted for Log X Model')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.show()
```

## Normality and Independence of Residuals


```python
# Normality of Residuals: Shapiro-Wilk test
from scipy.stats import shapiro

# Perform the Shapiro-Wilk test on the model residuals
shapiro_test = shapiro(log_X_residuals)
```

```python
# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# Histogram of the residuals
sns.histplot(log_X_residuals, kde=True, ax=ax[0])
ax[0].axvline(x=0, color='red', linestyle='--')
ax[0].set_title('Histogram of Residuals')
ax[0].set_xlabel('Residuals')
ax[0].set_ylabel('Frequency')

# Q-Q plot of the residuals
stats.probplot(log_X_residuals, dist="norm", plot=ax[1])
ax[1].set_title('Q-Q Plot of Residuals')

plt.tight_layout()
plt.show()
```

```python
print(shapiro_test)
```

* The p-value is greater than the typical alpha level of 0.05, which suggests that we do not reject the null hypothesis. This indicates that the residuals could be considered **normally distributed**.

```python
# Durbin-Watson test for independence of residuals
from statsmodels.stats.stattools import durbin_watson

# Perform the Durbin-Watson test on the model residuals
dw_test = durbin_watson(log_X_residuals)
print(dw_test)
```

* The Durbin-Watson statistic ranges from 0 to 4, with values around 2 suggesting no autocorrelation, values below 1 or above 3 suggesting strong autocorrelation, and values between 1 and 3 indicating a degree of autocorrelation.
* There is no strong evidence of autocorrelation in the residuals.

## Breusch-Pagan Test for homoscedasticity

```python
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.compat import lzip
import statsmodels.api as sm

# Breusch-Pagan test for homoscedasticity
bp_test = sm.stats.diagnostic.het_breuschpagan(log_X_residuals, model_log_X.model.exog)

# Organize the test output
bp_test_results = lzip(['Breusch-Pagan Test', 'p-value', 'f-value', 'f p-value'], bp_test)
bp_test_results
```

* Since the p-values are below the common alpha level of 0.05, this suggests that we can reject the null hypothesis of homoscedasticity, which indicates that **there is evidence of heteroscedasticity** in the residuals of the model with log-transformed independent variables.
* From the residuals plot, it appears that there is some pattern to the residualsthey are not randomly dispersed around the zero line, suggesting that the assumption of homoscedasticity might be violated. This is consistent with the results of the Breusch-Pagan test, which indicated evidence of heteroscedasticity.

## If the assumption of homoscedasticity is not met, there are several approaches you can take to address heteroscedasticity:

* *Robust Standard Errors*: This method adjusts the standard errors of the coefficients to account for the heteroscedasticity without altering the model itself. It's a common approach when dealing with heteroscedasticity.
* *Weighted Least Squares (WLS)*:This method involves giving less weight to observations that have larger variances, which can help in case the variances are a function of the predictors.

### Robust Standard Errors

```python
# Refit the OLS model using robust standard errors to address heteroscedasticity
model_robust = model_log_X.get_robustcov_results(cov_type='HC3')
```

```python
model_robust.summary()
```

* The robust standard errors have been used to calculate the t-statistics and corresponding p-values, which are designed to be valid even in the presence of heteroscedasticity.

```python
model_log_X.summary()
```

### WLS

* For the weights in the WLS model, a common approach is to use the inverse of the predicted variance from a model of the residuals. Since we don't have direct access to the individual variances of the residuals, we'll use the inverse of the squared residuals from the OLS model as an approximation for the weights.

```python
# # Calculate weights as the inverse of the squared residuals from the log-transformed OLS model
weights = 1 / (log_X_residuals ** 2)
```

```python
# Fit the WLS model using the weights
wls_model = sm.WLS(y, sm.add_constant(np.log(X)), weights=weights).fit()
```

```python
wls_model.summary()
```

```python
sm.stats.diagnostic.het_breuschpagan(wls_model.resid, wls_model.model.exog)
```

**Observations:**
* The extraordinarily high R-squared value and F-statistic suggest that the model might be too perfectly fit to the data, which could be a sign of **overfitting** or an artifact of the weighting scheme.
* The use of squared residuals from the OLS model as weights in the WLS model assumes that the variance of the residuals is proportional to the square of the observed value, which may not always be the case.
* p-values lower than 5% mean that the model is still exhibiting signs of heteroscedasticity.
